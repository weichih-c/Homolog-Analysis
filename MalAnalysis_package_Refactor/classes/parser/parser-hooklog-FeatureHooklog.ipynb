### Dot Matrix ###

```python
def apiEqual(x, y):
    
    #if x[1][0] == y[1][0] and x[1][1] == y[1][1]:
    if x[1] == y[1]:
        return True
    else:
        return False
```

```python
import pylab
import matplotlib.pyplot as plt
%matplotlib inline

# calc: Dot Matrix
def dotMatrix(fh1, fh2, window):
    x = []
    y = []

    for i in range(len(fh1)):
        for j in range(len(fh2)):
            if apiEqual(fh1[i], fh2[j]):
                x.append(i)
                y.append(j)
                
    return x, y

def saveDotMatrix(x, y, fh1, fh2, fwindow):
    
    orig_size = pylab.rcParams['figure.figsize']
    pylab.rcParams['figure.figsize'] = (8, 8)

    plt.clf()
    long_api = len(fh2) if len(fh2) > len(fh1) else len(fh1)
    pylab.scatter(x,y, s=1)
    pylab.xlim(0, long_api)
    pylab.ylim(0, long_api)
    pylab.xlabel("%s (length %i)" % (fh1.digitname, len(fh1)))
    pylab.ylabel("%s (length %i)" % (fh2.digitname, len(fh2)))
    pylab.title("Dot matrix (window size %i)" % window)
    plt.tight_layout()
    #plt.savefig('output/dotmatrix/' + 'dotmatrix_' + fh1.digitname + '_' + fh2.digitname + '.pdf', dpi=300)
    
    pylab.rcParams['figure.figsize'] = orig_size
```
